```python
import zipfile
import os
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.remote.command import Command
import time
import json
```

```python
#definimos el driver del navegador a utilizar
driver = webdriver.Chrome('chromedriver.exe')
#definimos el link a scrapear
pagina=driver.get('http://iinei.inei.gob.pe/iinei/siemweb/sist/frame.asp?idaplicacion=003')
#cambiamos de frame
driver.switch_to.frame('main')
#cambiamos de frame
driver.switch_to.frame('Izquierda')
#buscamos elemento por un texto parcial
driver.find_element_by_partial_link_text('Sistema de Índices de Precios').click()
#damos 1 segundo de espera para evitar posibles rechazo en peticiones a enviar
time.sleep(1)
driver.find_element_by_partial_link_text('Al por Mayor').click()
time.sleep(1)
driver.find_element_by_partial_link_text('Índice General').click()
#volvemos al contenido por defecto de la pagina para revisar otro frame
driver.switch_to.default_content()
driver.switch_to.frame('main')
driver.switch_to.frame('Superior')
time.sleep(1)
driver.find_element_by_name('chkseleccion').click()
time.sleep(1)
#buscamos elemento por su xpath
driver.find_element_by_xpath('/html/body/table/tbody/tr/td/table/tbody/tr[7]/td/form/table/tbody/tr[5]/td/input').click()
time.sleep(1)
#definimos una lista de opciones como elemento
elemento=driver.find_element_by_name('left_select')
#revisaremos las opciones dentro de elemento
for option in elemento.find_elements_by_tag_name('option'):
#si la opcion se ajusta al texto definido, se hace click en ella
    if option.text == 'Índice General al Por Mayor (M/Índice Base 1994=100,0)':
        option.click() # select() in earlier versions of webdriver
        break
time.sleep(1)

driver.find_element_by_xpath('.//table/tbody/tr/td/table/tbody/tr[7]/td/form/table/tbody/tr[3]/td[3]/img').click()
time.sleep(1)
driver.find_element_by_xpath('/html/body/table/tbody/tr/td/table/tbody/tr[7]/td/form/table/tbody/tr[7]/td/table/tbody/tr/td[2]/input').click()
time.sleep(1)
driver.find_element_by_xpath('.//table/tbody/tr/td/table/tbody/tr[6]/td/table/tbody/tr[4]/td[2]/img').click()
cuadro=driver.find_elements_by_xpath('.//table/tbody/tr/td/table/tbody/tr[6]/td/table/tbody/tr[3]/td/table/tbody')
#Obtenemos los elementos de un cuadro
def parseacuadro():
    cuadrito=[]
    for i in range(17):
        i=str(i)
        for tr in driver.find_elements_by_xpath('/html/body/table/tbody/tr/td/table/tbody/tr[6]/td/table/tbody/tr[3]/td/table/tbody/tr['+i+']'):
            tds = tr.find_elements_by_tag_name('td')
            cuadrito.append( [td.text for td in tds])
    return cuadrito
tabla=parseacuadro()
print(tabla)
```

```python
#pasamos a formato json el resultado y creamos indentacion para mejor visualización
son_string = json.dumps(tabla, indent=3)
print(son_string)
```
